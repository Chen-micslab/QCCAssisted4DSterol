```python
from ML.tabularRegressionML import TabularRegressionML
import pandas as pd
import numpy as np
QM_before_lasso_data = pd.read_csv('./database/ML/QM_Before_Lasso.csv', index_col=0)
Standard_before_lasso_data = pd.read_csv('./database/ML/ST_Before_Lasso.csv', index_col=0,)
QMStandard_before_lasso_data_before_lasso_data = pd.read_csv('./database/ML/QMST_Before_Lasso.csv', index_col=0)
All_Sterol_before_lasso_data = pd.read_csv('./database/6_ALLFingerPrintDropDuplicates.csv', index_col=0, low_memory=False)
```

```python
# ---------------------Lasso hyperparameter search press num 8 is lasso--------------------------
myML = TabularRegressionML(QM_before_lasso_data, './ML/MLResult', seed_start=0, seed_stop=100,
                           seed_step=1, test_ratio=0.2, score=['rmse', 'r2', 'mae', 'percent_madae', 'percent_mae'],
                           cv_number=5)  # use autoML start machine learning
myML.analyze_tabular_data(targetIndex=0, featureIndexList=[[1, 3000]])
myML.grid = [
    {'regressor__alpha': [0.001,0.01,0.1,1,10,100]
     }]
myML.model_selection_input()
for seed in myML.seed_range:
    myML.start_grid_search(seed=seed)
myML.save_result_file(test_verbose=True, time_char=False, additive_name_char='QM_feature_selection')
```

```python
# ---------------------Feature select--------------------------
myML = TabularRegressionML(QM_before_lasso_data, './ML/MLResult', seed_start=0, seed_stop=1,
                           seed_step=1, test_ratio=0.2, score=['rmse', 'r2', 'mae', 'percent_madae', 'percent_mae'],
                           cv_number=5)  # use autoML start machine learning
myML.analyze_tabular_data(targetIndex=0, featureIndexList=[[1, 3000]])
myML.grid = [
    {'regressor__alpha': [1]
     }]
myML.set_seed_range(seed_start= 0, seed_stop=1, seed_step=1)
myML.model_selection_input()
for seed in myML.seed_range:
    myML.start_grid_search(seed=seed)
coef = myML.best_estimator['regressor'].coef_
coef = np.insert(coef, 0, 1).astype(bool)
QM_after_lasso_data = QM_before_lasso_data.loc[:,coef]
print(f'selected feature {QM_after_lasso_data.columns}')
QM_after_lasso_data.to_csv('./database/ML/QM_After_Lasso.csv')

myML.clear_output_file()
```

```python
# -------------------Use standard sterol ccs value as out-of-sample validation-----------------------------
# ------------------------Grid search of SVR by 3 steps---------------------
QM_after_lasso_data = pd.read_csv('./database/ML/QM_After_Lasso.csv', index_col=0)
myML = TabularRegressionML(QM_after_lasso_data, './ML/MLResult', seed_start=0, seed_stop=100,
                           seed_step=1, test_ratio=0.2, score=['rmse', 'r2', 'mae', 'percent_madae', 'percent_mae'],
                           cv_number=5)  # use autoML start machine learning
myML.analyze_tabular_data(targetIndex=0, featureIndexList=[[1, 3000]])
myML.grid = [{
    'regressor__kernel': ['rbf'],
    'regressor__C': [0.001,0.01,0.1,100,300,500],
    'regressor__gamma': [0.001,0.01,0.1,1]
}]
myML.model_selection_input()
for seed in myML.seed_range:
    myML.start_grid_search(seed=seed)
myML.save_result_file(time_char=False,additive_name_char='QM_First_Search')
myML.clear_output_file()
```
